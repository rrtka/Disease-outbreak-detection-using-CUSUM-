```python
data = data[['YEAR', 'WEEK', '% WEIGHTED ILI', 'REGION TYPE']]
data.columns = ['Year', 'Week', 'Flu', 'Region']
```

```python
data['Flu'] = pd.to_numeric(data['Flu'], errors='coerce')
data = data.dropna()
data = data.reset_index(drop=True)
```

```python
data['Time'] = data['Year'].astype(int).astype(str) + "-W" + data['Week'].astype(int).astype(str)
data.shape
data.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(data['Flu'])
plt.axhline(y=data['Flu'].mean(), linestyle='--', label='Mean')
plt.title("Weekly Flu Trend")
plt.xlabel("Time (Weeks)")
plt.ylabel("Flu (%)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
flu = data['Flu']
mean = flu.mean()
std = flu.std()
k = 0.5 * std
h = 3 * std
cusum = [0]
signals = []
for i in range(1, len(flu)):
    val = max(0, cusum[i-1] + (flu.iloc[i] - mean - k))
    cusum.append(val)    
    if val > h:
        signals.append(i)
```

```python
plt.figure(figsize=(14,6))
plt.plot(flu, label='Flu Level')
plt.scatter(signals, flu.iloc[signals], color='red', label='Outbreak')
plt.axhline(y=mean, linestyle='--', label='Mean')
plt.axhline(y=mean + 3*std, linestyle='--', label='Threshold')
plt.xticks(range(0, len(data), 50), rotation=45)
plt.title("Disease Outbreak Detection using CUSUM")
plt.xlabel("Time (Weeks)")
plt.ylabel("Flu (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("Total Outbreak Signals:", len(signals))
```
